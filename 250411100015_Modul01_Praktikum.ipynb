# 3. Buat kode program yang meminta input suhu dalam satuan Fahrenheit, lalu konversi suhu tersebut ke dalam satuan Celcius & Kelvin, seperti pada gambar berikut.
nomer 3.png

```python
fahrenheit = float(input("masukkkan suhu dalam satuan fahrenheit :"))
celcius = (fahrenheit-32)*5/9
kelvin = (fahrenheit-32)*5/9+273.15
print ("fahrenheit -> celcius :", celcius)
print ("fahrenheit -> kelvin  :", kelvin)
```
