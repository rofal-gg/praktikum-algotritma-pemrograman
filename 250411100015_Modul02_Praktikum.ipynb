# 1. Buatlah code untuk menentukan Nilai grade dari Nilai akhir Mahasiswa dengan ketentuan sebagai berikut :
- Input berupa Nilai akhir
- Output berupa Nilai grade
- Nilai 90 – 100 => A
- Nilai 80 – 89 => B
- Nilai 70 – 79 => C
- Nilai 60 – 69 => D
- Nilai 50 – 59 => E
- Nilai >50 => F

```python
nilai=int(input("masukkan nilai: "))
if nilai>=90:
  print ("grade: A")
elif nilai>=80:
  print ("grade: B")
elif nilai>=70:
  print ("grade: C")
elif nilai>=60:
  print ("grade: D")
elif nilai>=50:
  print ("grade: E")
else:
  print ("grade: F")
```

# 2. Buatlah code untuk mengubah standar waktu Indonesia yang 24 jam menjadi 12 jam,
contoh 18.30 WIB menjadi 6.30 p.m. dengan ketentuan sebagai berikut :
- Input user berupa format 24 jam
- Output berupa format 12 jam

```python
print ("konversi waktu format 24jam ke 12 jam")
print ("masukkan jam dan menit")
jam=int(input("masukkan jam: "))
menit=int(input("masukkan menit: "))
print ("jam yang kamu masukkan adalah: ",jam,".",menit,"WIB")
if jam==0:
  print (jam,".",menit,"WIB menjadi:",jam,".",menit,"a.m.")
elif jam<12:
  print (jam,".",menit,"WIB menjadi:",jam,".",menit,"a.m.")
elif jam==12:
  print (jam,".",menit,"WIB menjadi:",jam,".",menit,"p.m.")
else:
  print (jam,".",menit,"WIB menjadi:",jam-12,".",menit,"p.m.")


```
