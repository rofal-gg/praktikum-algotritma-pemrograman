# **1. Soal 1 (20 Poin)**
Buatlah program yang meminta *user* memasukkan sebuah **bilangan bulat positif**.
Program kemudian akan menentukan apakah bilangan tersebut merupakan bilangan
prima atau bukan (Hint: Bilangan prima merupakan bilangan yang hanya memiliki
dua (2) faktor, seperti 2, 3, 5, 7, 11, ...).

**Ketentuan:**
- Gunakan perulangan *for/while* **untuk mencari faktor dari bilangan bulat positif** yang telah **dimasukkan oleh user (sebagai input)**.
- Gunakan statement break jika faktor yang ditemukan lebih dari dua (2).
- Jika bilangan yang dimasukkan oleh *user* adalah bilangan prima, maka
tampilkan pesan “Angka ... merupakan bilangan prima”. Jika bukan, maka
tampilkan pesan “Angka ... bukanlah bilangan prima”.

**Contoh output:**

image.png

```python
angka = int(input("masukkan bilangan bulat positif: "))
if angka < 1:
    print("angka tidak valid")
elif angka == 2:
    print(f"angka {angka} merupakan bilangan prima")
elif angka % 2 == 0 or angka == 1:
    print(f"angka {angka} bukanlah bilangan prima")
else:
    faktor = 0
    for i in range (1, angka + 1):
        if angka % i == 0:
            faktor += 1
            if faktor > 2:
                print(f"angka {angka} bukanlah bilangan prima")
                break
    if faktor == 2:
        print(f"angka {angka} merupakan bilangan prima")
```

# 2. **Soal 2 (30 Poin)**
Buatlah program yang meminta *user* untuk memasukkan angka bulat (integer)
sebanyak **n**. Setelah semua angka dimasukkan, program harus:
- Menampilkan jumlah total dari 10 angka tersebut.
- Menampilkan frekuensi berapa kali
**angka genap** dimasukkan.

**Ketentuan:**
- Gunakan perulangan *for loop* untuk mengontrol **_n_** masukan/input.
- Gunakan Operator *Assignment* untuk menghitung total angka.
- Gunakan *Branching* di dalam loop untuk mengecek angka genap.

**contoh output:**

image.png

```python
n = int(input("anda ingin berapa kali input?\t: "))
print()
total = 0
genap = 0
for i in range (1, n+1):
    angka = int(input(f"masukkan angka ke-{i}\t\t: "))
    total += angka
    if angka % 2 == 0:
        genap += 1
print()
print(f"total keseluruhan\t\t: {total}")
print(f"frekuensi angka genap\t\t: {genap}")
```

# **3. Soal 3 (50 Poin)**
Buatlah program kasir sederhana yang meminta user memasukkan harga produk satu
per-satu, dan proses ini akan terus berjalan hingga *user* memasukkan “0” sebagai
harga barang. Setelah proses *input* harga selesai, program harus:
- Menghitung & menampilkan **total belanja**.
- Meminta user memasukkan **jumlah uang yang dibayarkan** (uang tunai).
- Menghitung dan menampilkan **kembalian** *user*. Jika uang yang dibayarkan
kurang dari total belanja, cetak pesan “Uang tidak cukup, masukkan ulang
nominal”, lalu ulangi proses ini hingga uang yang dibayarkan oleh user cukup.

**Ketentuan:**
- Gunakan perulangan *while loop* dengan nilai 0 sebagai penanda berhenti loop.
- Gunakan Operator Assignment untuk menghitung total belanja.
- Gunakan branching untuk mengecek apakah uang *user* cukup atau tidak.
- Gunakan perulangan *while loop* untuk mengulangi proses pembayaran jika
uang *user* tidak cukup.

**Contoh output:**

image.png

```python
total=0
while True:
    harga = int(input("masukkan harga barang\t\t: "))
    if harga < 0:
        print("harga tidak valid")
    elif harga == 0:
        break
    else:
        total += harga
print()
print(f"total belanja\t\t\t: {total}")
print()
while True:
    bayar=int(input("jumlah uang yang dibayarkan\t: "))
    if bayar < total:
        print("uang tidak cukup, masukkan ulang nominal")
        print()
    else:
        print(f"uang kembalian\t\t\t: {bayar - total}")
        break
```
