# 1. SOAL 1 (Poin: 20)
Poligon beraturan adalah poligon bersisi n yang dimana semua sisinya sama panjang,
dan semua sudutnya memiliki derajat yang sama (yaitu, poligonnya sama sisi dan
sama sudut). Luas dari poligon beraturan adalah sebagai berikut (yang dimana s
adalah panjang sisi dari poligon beraturan, dan n merupakan banyak sisi):

image.png

Buatlah program yang meminta pengguna untuk memasukkan jumlah sisi dan panjang
sisi tersebut dari suatu poligon beraturan dan tampilkan luasnya. Jika hasil adalah
desimal, maka jumlah maksimal angka dibelakang koma yang ditampilkan adalah 2.
Berikut ini adalah contoh tampilan console-nya ketika program dijalankan:

image.png


```python
n= int(input("masukkan jumlah sisi poligon\t: "))
s= float(input("masukkan panjang sisi poligon\t: "))
import math
area=(n*(math.pow(s, 2)))/(4*(math.tan(math.pi/n)))
print(f"Luas poligon adalah\t\t: {area:.2f}")
```

Buatlah program untuk menghitung **jarak total** dari tiga titik koordinat, yaitu A(x1,y1), B(x2, y2), dan C(x3, y3).

Program harus:
- Meminta input koordinat (x1, y1, x2, y2, x3, y3)
- Menghitung jarak:
  - Antara A & B  d1
  - Antara B & C  d2
  - Total jarak = d1 + d2
- Gunakan rumus:

image.png

- Jika hasil adalah desimal, maka jumlah maksimal angka dibelakang koma
yang ditampilkan adalah 2.

Berikut ini adalah contoh tampilan console-nya ketika program dijalankan

image.png


```python
x_1= int(input("masukkan titik A - x1\t: "))
y_1= int(input("masukkan titik A - y1\t: "))
print()
x_2= int(input("masukkan titik A - x2\t: "))
y_2= int(input("masukkan titik A - y2\t: "))
print()
x_3= int(input("masukkan titik A - x3\t: "))
y_3= int(input("masukkan titik A - y3\t: "))
print()
import math
a= x_2-x_1
b= y_2-y_1
d_1= math.sqrt((math.pow(a,2))+ (math.pow(b,2)))
e= x_3-x_2
f= y_3-y_2
d_2= math.sqrt((math.pow(e,2))+ (math.pow(f,2)))
print (f"jarak A --> B\t\t: {d_1:.2f}")
print (f"jarak B --> C\t\t: {d_2:.2f}")
print (f"total jarak\t\t: {d_1+d_2:.2f}")
```

# 3. SOAL 3 (Poin: 50)
Jarak lingkaran besar adalah jarak antara dua titik pada permukaan bumi. Jika (x1, y1)
dan (x2, y2) menjadi garis lintang dan garis bujur geografis dari dua titik. Jarak
lingkaran besar antara dua titik dapat dihitung dengan menggunakan rumus berikut:

d =radius x arccos(sin(x1) x sin (x2) + cos(x1) cos(x2) x cos(y1 - y2))

Buatlah program yang meminta pengguna untuk memasukkan garis lintang dan garis
bujur dari dua titik di bumi dalam derajat, dan tampilkan jarak lingkaran besarnya.
Jari-jari bumi rata-rata adalah 6.371,01 km. Perhatikan bahwa anda perlu mengubah
derajat menjadi radian menggunakan fungsi **math.radian()** karena **fungsi trigonometri Python menggunakan radian.** Derajat lintang dan bujur dalam rumus
adalah untuk utara dan barat. Gunakan negatif untuk menunjukkan derajat selatan dan
timur.

Berikut ini adalah contoh tampilan console-nya ketika program dijalankan:

image.png


```python
x_1=float(input("masukkan garis lintang (x1)\t: "))
y_1=float(input("masukkan garis bujur (y1)\t: "))
print()
x_2=float(input("masukkan garis lintang (x2)\t: "))
y_2=float(input("masukkan garis bujur (y2)\t: "))
radius=6371.01
import math
a=math.radians(x_1)
b=math.radians(y_1)
c=math.radians(x_2)
d=math.radians(y_2)
d = radius*math.acos(math.sin(a)*math.sin(c)+math.cos(a)*math.cos(c)*math.cos(b-d))
print(f"jarak antara kedua titik\t: {d}")
```
